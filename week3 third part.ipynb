```python
import pandas as pd
```

```python
url = "https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M"
```

```python
table_list = pd.read_html(url)
```

```python
df = table_list[0]  #three tables, first one is needed
df
```

# clean dataframe

```python
df2=df.loc[df["Borough"]!="Not assigned",:]
df2.reset_index(drop=True,inplace=True)
df2
```

# combine Neighbourhod

```python
new_df = pd.DataFrame()
uni_post = df2.iloc[:,0].unique()
for code in uni_post:
    tem_df = df2[df2.iloc[:,0]==code]
    if(len(tem_df)>1):
#         print(tem_df)
        combi = tem_df.iloc[:,2].values
        str1 = ", "
        tem_df = pd.DataFrame(data=[[tem_df.iloc[0,0],tem_df.iloc[0,1],str1.join(combi)]],columns=tem_df.columns.values)
    new_df = pd.concat([new_df,tem_df])
new_df.reset_index(drop=True,inplace=True)
new_df
# new_df.iloc[:,2]=new_df.iloc[:,2]
    
```

```python
import numpy as np # library to handle data in a vectorized manner

import pandas as pd # library for data analsysis
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)

import json # library to handle JSON files

# !conda install -c conda-forge geopy --yes # uncomment this line if you haven't completed the Foursquare API lab
from geopy.geocoders import Nominatim # convert an address into latitude and longitude values

import requests # library to handle requests
from pandas.io.json import json_normalize # tranform JSON file into a pandas dataframe

# Matplotlib and associated plotting modules
import matplotlib.cm as cm
import matplotlib.colors as colors
import matplotlib.pyplot as plt
# import k-means from clustering stage
from sklearn.cluster import KMeans

#!conda install -c conda-forge folium=0.5.0 --yes # uncomment this line if you haven't completed the Foursquare API lab
import folium # map rendering library

print('Libraries imported.')
```

```python
address = 'City in Ontario, Canada'

geolocator = Nominatim(user_agent="ca_explorer")
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
print('The geograpical coordinate of Toronto are {}, {}.'.format(latitude, longitude))
```

```python
# create map of New York using latitude and longitude values
map_toronto = folium.Map(location=[latitude, longitude], zoom_start=10)

# add markers to map
for lat, lng, borough, neighborhood in zip(neighborhoods['Latitude'], neighborhoods['Longitude'], neighborhoods['Borough'], neighborhoods['Neighbourhood']):
    label = '{}, {}'.format(neighborhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_toronto)  

map_toronto
```

```python
data = neighborhoods[['Latitude','Longitude']]
data = data.values
```

```python
k_means = KMeans(init="k-means++", n_clusters=4, n_init=12)
k_means.fit(data)
```

```python
k_means_labels = k_means.labels_
k_means_labels
```

```python
k_means_cluster_centers = k_means.cluster_centers_
k_means_cluster_centers
```

```python
# initialize the plot with the specified dimensions.
fig = plt.figure(figsize=(15, 10))

# colors uses a color map, which will produce an array of colors based on
# the number of labels. We use set(k_means_labels) to get the
# unique labels.
colors = plt.cm.Spectral(np.linspace(0, 1, len(set(k_means_labels))))

# create a plot
ax = fig.add_subplot(1, 1, 1)

# loop through the data and plot the datapoints and centroids.
# k will range from 0-3, which will match the number of clusters in the dataset.
for k, col in zip(range(len([[4,4], [-2, -1], [2, -3], [1, 1]])), colors):

    # create a list of all datapoints, where the datapoitns that are 
    # in the cluster (ex. cluster 0) are labeled as true, else they are
    # labeled as false.
    my_members = (k_means_labels == k)
    
    # define the centroid, or cluster center.
    cluster_center = k_means_cluster_centers[k]
    
    # plot the datapoints with color col.
    ax.plot(data[my_members, 0], data[my_members, 1], 'w', markerfacecolor=col, marker='.',markersize=15)
    
    # plot the centroids with specified color, but with a darker outline
    ax.plot(cluster_center[0], cluster_center[1], 'o', markerfacecolor=col,  markeredgecolor='k', markersize=10)

# title of the plot
ax.set_title('KMeans(large cirle is centroid, small circle is datapoint)')

# remove x-axis ticks
ax.set_xticks(())

# remove y-axis ticks
ax.set_yticks(())

# show the plot
plt.show()
```
